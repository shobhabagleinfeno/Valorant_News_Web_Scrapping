```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
valo_news=requests.get('https://playvalorant.com/en-gb/news/')
print(valo_news)
```

```python
soup=BeautifulSoup(valo_news.content,'html.parser')
print(soup.prettify())
```

```python
news=soup.find_all('div')
```

```python
news_data=[]
image=soup.find_all('div',class_='innerWrapper')
title=soup.find_all('div',class_='sc-bad9cda9-2 XONNe')
date=soup.find_all('div',class_='sc-bad9cda9-3 biqGyb')
headline=soup.find_all('div',class_='sc-ce9b75fd-0 lmZfRs')
summary=soup.find_all('div',class_='sc-4225abdc-0 lnNUuw')

for img, title, date, headline, summary in zip(image, title, date, headline, summary):
    img_tag = img.find('img')
    image_url = img_tag.get('src') if img_tag else "No Image Found"
    news_data.append({
        'Image': image_url,
        'Title': title.text.strip(),
        'Date': date.text.strip(),
        'Headline': headline.text.strip(),
        'Summary': summary.text.strip()
    })


with open('valo_news.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['Image', 'Title', 'Date', 'Headline', 'Summary'])
    writer.writeheader()
    for n in news_data:
        writer.writerow(n)
```
